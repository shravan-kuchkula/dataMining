## How to rename columns ?

Let's pretend that we want to rename the `Colors Reported` and `Shape Reported` so that they don't have a space in them.

We are going to use the dataframe's `rename` method.

```python
ufo.rename(columns = {'Colors Reported':'Colors_Reported', 'Shape Reported':'Shape_Reported'}, inplace=True)
```

```python
ufo.columns
```

So, this is one way of doing this. Let's see another way!

First let's create a python list of strings

```python
ufo_cols = ['city', 'colors reported', 'shape reported', 'state', 'time']
```

```python
ufo.columns = ufo_cols
```

```python
ufo.head()
```
